```python
from bq_helper import BigQueryHelper
```

```python
#BigQueryHelper requires creating one object (2 arguments: project name & dataset name)) Here we make a new one
bq_assistant=BigQueryHelper('bigquery-public-data','openaq')
```

```python
bq_assistant.list_tables()
```

```python
# Now we are doing some basic EDA
bq_assistant.head('global_air_quality',num_rows=3)
```

```python
# have a overview of columns
bq_assistant.table_schema('global_air_quality')
```

```python
# After have an overlook of dataset, we can set up a query
QUERY = "SELECT location, timestamp, pollutant FROM `bigquery-public-data.openaq.global_air_quality`"
```

```python
# it is a good habit to always check how big the dataset is that you will execute scanning or querying
bq_assistant.estimate_query_size(QUERY)
```

```python
# before we run a query, it is another good habit to get results into a pandas dataframe
df=bq_assistant.query_to_pandas(QUERY)
```
